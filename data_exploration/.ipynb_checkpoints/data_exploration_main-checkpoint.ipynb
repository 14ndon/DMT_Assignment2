# Imports

## Import Libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

# Configure visualizations
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline
```

## Import Data

```python
# loading data
df = pd.read_csv('../data/training_set_VU_DM.csv')
df['date_time'] = pd.to_datetime(df['date_time'])
df.head()
```

## Visualizations

```python
# Check the distribution of clicks 
click_distribution = train['click_bool'].value_counts(normalize=True)
click_distribution.plot(kind='bar')
plt.title('Distribution of Clicks vs No Clicks')
plt.xlabel('Clicked')
plt.ylabel('Proportion')
plt.show()
```

```python
# Compare count of bookings for each click_bool value
booking_count = df.groupby('click_bool')['booking_bool'].value_counts().unstack(fill_value=0)

# Plot the comparison (bars side by side)
booking_count.plot(kind='bar', color=['blue', 'lightblue'], figsize=(8, 6))
plt.title('Count of Bookings by Click Status')
plt.xlabel('Click Status (0 = Not Clicked, 1 = Clicked)')
plt.ylabel('Count of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(['No Booking', 'Booking'])
plt.show()
```

```python
df_clicked = df[df['click_bool'] == 1]

# Count the occurrences of booking_bool (0 vs 1) for click_bool == 1
booking_count = df_clicked['booking_bool'].value_counts()

# Plot the comparison
booking_count.plot(kind='bar', color=['blue', 'lightblue'], figsize=(8, 6))
plt.title('Booking Count for Clicked Results (click_bool = 1)')
plt.xlabel('Booking Status (0 = No Booking, 1 = Booking)')
plt.ylabel('Count of Bookings')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(['No Booking', 'Booking'])
plt.show()
```

```python
# Calculate the percentage of promotions for booked vs non-booked
booking_promotions = df.groupby('booking_bool')['promotion_flag'].mean() * 100

# Calculate the percentage of promotions for clicked vs non-clicked
click_promotions = df.groupby('click_bool')['promotion_flag'].mean() * 100

# Create the plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for booking vs non-booking promotions
axes[0].bar(booking_promotions.index.astype(str), booking_promotions, color=['blue', 'lightblue'])
axes[0].set_title('Promotion Rate for Booked vs Non-Booked')
axes[0].set_xlabel('Booking Status (0 = Not Booked, 1 = Booked)')
axes[0].set_ylabel('Percentage of Promotions')
axes[0].set_ylim(0, 40)
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['Not Booked', 'Booked'])

# Plot for clicked vs non-clicked promotions
axes[1].bar(click_promotions.index.astype(str), click_promotions, color=['blue', 'lightblue'])
axes[1].set_title('Promotion Rate for Clicked vs Non-Clicked')
axes[1].set_xlabel('Click Status (0 = Not Clicked, 1 = Clicked)')
axes[1].set_ylabel('Percentage of Promotions')
axes[1].set_ylim(0, 40)
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Not Clicked', 'Clicked'])

plt.tight_layout()
plt.show()
```

```python
# Group by promotion and calculate mean stay length
promo_stay_length = df.groupby('promotion_flag')['srch_length_of_stay'].mean()

# Plot
plt.figure(figsize=(8, 6))
plt.bar(promo_stay_length.index.astype(str), promo_stay_length, color=['blue', 'lightblue'])
plt.title('Average Stay Length with and without Promotion')
plt.xlabel('Promotion Flag (0 = No Promotion, 1 = Promotion)')
plt.ylabel('Average Stay Length (Days)')
plt.xticks([0, 1], ['No Promotion', 'Promotion'])
plt.show()
```

```python
# Calculate the log price
log_price = np.log(df['price_usd'])

# Plot the log price for different lengths of stay
plt.figure(figsize=(8, 6))
plt.scatter(df['srch_length_of_stay'], log_price, alpha=0.6)
plt.title('Log Price vs Length of Stay')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Log of Price')
plt.show()
```

```python
# Extract month from date
df['month'] = df['date_time'].dt.month

# Calculate booking rate per month
booking_rate_per_month = df.groupby('month')['booking_bool'].mean() * 100

# Plot
plt.figure(figsize=(8, 6))
booking_rate_per_month.plot(kind='line', marker='o')
plt.title('Booking Rate Over Time (by Month)')
plt.xlabel('Month')
plt.ylabel('Booking Rate (%)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Group by customer country and calculate booking percentage
booking_by_country = df.groupby('visitor_location_country_id')['booking_bool'].mean() * 100

# Plot
plt.figure(figsize=(8, 6))
booking_by_country.plot(kind='bar', color=['blue' if country != 219 else 'red' for country in booking_by_country.index])
plt.title('Booking Percentage by Customer Country')
plt.xlabel('Customer Country ID')
plt.ylabel('Booking Percentage (%)')
plt.show()
```

```python
# Group by customer country and calculate booking percentage
booking_by_country = df.groupby('visitor_location_country_id')['booking_bool'].mean() * 100

# Identify the country with the highest booking rate
highest_booking_country = booking_by_country.idxmax()
highest_booking_rate = booking_by_country.max()

# Plot
plt.figure(figsize=(8, 6))
# Use a color map to highlight the highest booking country
colors = ['blue' if country != highest_booking_country else 'red' for country in booking_by_country.index]
booking_by_country.plot(kind='bar', color=colors)
plt.title(f'Booking Percentage by Customer Country (Highest: Country {highest_booking_country} - {highest_booking_rate:.2f}%)')
plt.xlabel('Customer Country ID')
plt.ylabel('Booking Percentage (%)')

# Make the x-axis labels more readable
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by hotel country and calculate booking probability
booking_prob_by_hotel_country = df.groupby('prop_country_id')['booking_bool'].mean()

# Plot
plt.figure(figsize=(8, 6))
booking_prob_by_hotel_country.plot(kind='bar', color=['blue'])
plt.title('Booking Probability by Hotel Country ID')
plt.xlabel('Hotel Country ID')
plt.ylabel('Booking Probability')
plt.show()
```

```python
# Calculate the correlation between hotel_country_id and booking rate
booking_prob_by_hotel_country.corr(df['booking_bool'])
```

```python
# Filter only booked results
df_booked = df[df['booking_bool'] == 1]

# Plot: Price vs. Star Rating
plt.figure(figsize=(8, 6))
plt.scatter(df_booked['prop_starrating'], df_booked['price_usd'], color='purple')
plt.title('Price vs Star Rating for Booked Hotels')
plt.xlabel('Star Rating')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Plot price distribution by star rating
plt.figure(figsize=(8, 6))
df.groupby('prop_starrating')['price_usd'].mean().plot(kind='bar', color='orange')
plt.title('Average Price by Star Rating')
plt.xlabel('Star Rating')
plt.ylabel('Average Price (USD)')
plt.show()
```

```python
# Extract day, month, and year from date_time
df['day'] = df['date_time'].dt.day
df['month'] = df['date_time'].dt.month
df['year'] = df['date_time'].dt.year

# Plot: Distribution of Searches by Month
plt.figure(figsize=(8, 6))
df['month'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Searches by Month')
plt.xlabel('Month')
plt.ylabel('Number of Searches')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

```python
# Calculate the booking rate for each star rating
star_rating_booking_rate = df.groupby('prop_starrating')['booking_bool'].mean()

# Create a line plot to show the relationship
plt.figure(figsize=(8, 6))
sns.lineplot(x=star_rating_booking_rate.index, y=star_rating_booking_rate.values, marker='o')
plt.title("Booking Rate by Star Rating")
plt.xlabel("Hotel Star Rating")
plt.ylabel("Average Booking Rate")
plt.xticks(range(1, 6))  # Assuming star ratings are between 1 and 5
plt.show()
```

```python
# Group by length of stay and calculate mean booking rate
df_length_of_stay = df.groupby('srch_length_of_stay')['booking_bool'].mean().reset_index()

# Plot the booking rate by length of stay
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_length_of_stay, x='srch_length_of_stay', y='booking_bool', marker='o')
plt.title("Average Booking Rate vs Length of Stay")
plt.xlabel("Length of Stay (days)")
plt.ylabel("Average Booking Rate")
plt.show()
```

```python
# Define the threshold price as the median of 'price_usd'
price_threshold = df['price_usd'].median()

# Create a new column to categorize properties as high or low priced
df['price_category'] = ['High' if x > price_threshold else 'Low' for x in df['price_usd']]

# Group by price category and promotion flag to calculate mean booking rate
df_promotion_booking = df.groupby(['price_category', 'promotion_flag'])['booking_bool'].mean().reset_index()

# Plot the results
plt.figure(figsize=(8, 6))
sns.barplot(data=df_promotion_booking, x='price_category', y='booking_bool', hue='promotion_flag')
plt.title("Booking Rate vs Promotion for High and Low Priced Properties")
plt.xlabel("Price Category (High vs Low)")
plt.ylabel("Average Booking Rate")
plt.show()
```
